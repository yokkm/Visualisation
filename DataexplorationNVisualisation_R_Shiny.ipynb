```r
library(shiny)
library(ggplot2)
library(leaflet)

#--- This should be where you located the CSV file 
myData <- read.csv('/Users/yokk/assignment-02-formatted-data.csv', sep = ",")
myData

#coral name into lists
mylist <- as.list(myData$type)
myData$bleaching_Numeric <- as.numeric (sub("%","",myData$bleaching))


#UIPAGE START HERE
ui <- fluidPage(
  verticalLayout(
    #level1
    selectInput("select", "Select Coral type:",
                choices = as.character(unique(myData$type)), selected = TRUE),
    #level2
    #submitButton("Update coral type"),
    #level3
    leafletOutput("map", height="600px"),
    #level4
    h3(textOutput("caption")),
    actionButton("go","Plot"),
    #level5
    plotOutput("distPlot", height="300px")
  )#END OF VERTICAL LAYOUT
  
)#UIPAGE ENDS HERE

#-----------------------------------

#SERVER STARTS HERE
server <- function(input, output ,session) {
  
  
  
#only print existing data on the map  
  output$map <- renderLeaflet({
    if (input$select == "sea pens") {
      sdata <- myData[myData$type == input$select,]
      leaflet()%>% 
        setView(lng=131 , lat =-25, zoom=4) %>%
        addTiles(options = providerTileOptions(noWrap = TRUE)) %>% 
        addCircleMarkers(data=sdata,lng=~longitude, lat=~latitude,
                       layerId=~as.character(site), popup=~as.character(site),
                       radius=8 , color="yellow",  fillColor="red", stroke = TRUE,
                       fillOpacity = 0.8)
    }
    else if (input$select == "blue corals") {
      sdata <- myData[myData$type == input$select,]
      leaflet()%>% 
        setView(lng=131 , lat =-25, zoom=4) %>%
        addTiles(options = providerTileOptions(noWrap = TRUE)) %>% 
        addCircleMarkers(data=sdata,lng=~longitude, lat=~latitude,
                         layerId=~as.character(site), popup=~as.character(site),
                         radius=8 , color="yellow",  fillColor="red", stroke = TRUE,
                         fillOpacity = 0.8)
    }
    else if (input$select == "hard corals") {
      sdata <- myData[myData$type == input$select,]
      leaflet()%>% 
        setView(lng=131 , lat =-25, zoom=4) %>%
        addTiles(options = providerTileOptions(noWrap = TRUE)) %>% 
        addCircleMarkers(data=sdata,lng=~longitude, lat=~latitude,
                         layerId=~as.character(site), popup=~as.character(site),
                         radius=8 , color="yellow",  fillColor="red", stroke = TRUE,
                         fillOpacity = 0.8)
    }
    else if (input$select == "soft corals") {
      sdata <- myData[myData$type == input$select,]
      leaflet()%>% 
        setView(lng=131 , lat =-25, zoom=4) %>%
        addTiles(options = providerTileOptions(noWrap = TRUE)) %>% 
        addCircleMarkers(data=sdata,lng=~longitude, lat=~latitude,
                         layerId=~as.character(site), popup=~as.character(site),
                         radius=8 , color="yellow",  fillColor="red", stroke = TRUE,
                         fillOpacity = 0.8)
    }
    else{
      leaflet()%>% 
        setView(lng=131 , lat =-25, zoom=4) %>%
        addTiles(options = providerTileOptions(noWrap = TRUE)) %>% 
        addCircleMarkers(data=myData,lng=~longitude, lat=~latitude,
                         layerId=~as.character(site), popup=~as.character(site),
                         radius=8 , color="yellow",  fillColor="red", stroke = TRUE,
                         fillOpacity = 0.8)
      
    }
    })
  
  # create a reactive value that will store the click position
  data_of_click <- reactiveValues(clickedMarker=NULL)
  # store the click
  observeEvent(input$map_marker_click,{
    data_of_click$clickedMarker <- input$map_marker_click
  })
  

  
  formulaText <- reactive({
    paste('You select',input$select,' at ',data_of_click$clickedMarker$id,'-')
  })
  # Return the formula text for printing as a caption
  output$caption <- renderText({formulaText()})
  
  dataInput <- eventReactive(input$go, {
    
  a <- subset(myData, type == input$select)
  b <- subset(a, site==data_of_click$clickedMarker$id)
  })
  
   # c(input$go,data_of_click$clickedMarker$id)
  # Make a barplot depending of the selected point
  output$distPlot=renderPlot({
    
    print(ggplot(dataInput(), aes(x=year, y=bleaching_Numeric)) + 
            geom_bar(stat = "identity") +
            xlab("Year") + ylab("Bleaching Rate(%)"))
    
    
    
  })
  
  
  
  
  
  
  
}# End Server page here
shinyApp(ui = ui, server = server)
```
